#### Import numpy, matplotlib.pyplot and make it inline

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

To have a dataset or set of data points, the student must generate a pair of arrays <b>X</b> and <b>y</b> with the values in <b>X</b> equally distributed between <b>0</b> and <b>20</b> and the values in <b>y</b> such that: 
<b>yi = a*xi + b (and a = -2, b = 1)</b>


```python
x=np.random.uniform(0,20,100)
y=-2*x+1
```

```python
x=x.reshape(-1,1)
```

```python
new_x=np.append(np.ones((x.shape[0],1)),x,axis=1)
```

#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def BGD(x,y,learning_rate,n_epochs):
    Loss=[]
    theta0=[]
    theta1=[]
    lines=[]
    theta=np.random.rand(1,x.shape[1])
    for i in range(n_epochs):
        old_theta=theta.copy()
        print('--------------------epoch(',(i+1),')-----------------------')
        print(theta)
        theta-=learning_rate*(((x.dot(theta.T)-y.reshape(-1,1))*x).mean(axis=0))
        Loss.append(((x.dot(theta.T)-y.reshape(-1,1))**2).mean(axis=0))
        theta0.append(theta[0][0])
        theta1.append(theta[0][1])
        lines.append(new_x.dot(theta.T))
        check=theta==old_theta
        if check.all():
            break
    return theta,Loss,theta0,theta1,lines
```

#### Try your function with any input parameterss of your choice.

```python
theta,loss,theta0,theta1,lines=BGD(new_x,y,0.01,15)
```

#### Calculate r2 score

```python
pred_y=new_x.dot(theta.T)
```

```python
from sklearn.metrics import r2_score
print('R2: ',r2_score(pred_y,y))
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def MBGD(x,y,learning_rate,n_epochs,batch_size):
    Loss=[]
    theta0=[]
    theta1=[]
    lines=[]
    theta=np.random.rand(1,x.shape[1])
    for i in range(n_epochs):
        old_theta=theta.copy()
        print('--------------------epoch(',(i+1),')-----------------------')
        print(theta)
        size=len(x)
        perm=np.random.permutation(size)
        x_new=x[perm]
        y_new=y[perm]
        for it in range(0,size,batch_size):
            x_t=x_new[it:it+batch_size]
            y_t=y_new[it:it+batch_size]
            theta-=learning_rate*(((x_t.dot(theta.T)-y_t.reshape(-1,1))*x_t).mean(axis=0))
            Loss.append(((x_new.dot(theta.T)-y_new.reshape(-1,1))**2).mean(axis=0))
            theta0.append(theta[0][0])
            theta1.append(theta[0][1])
            lines.append(new_x.dot(theta.T))
        check=theta==old_theta
        if check.all():
            break
    return theta,Loss,theta0,theta1,lines
```

```python
theta,loss,theta0,theta1,lines=MBGD(new_x,y,0.001,500,32)
```

```python
pred_y=new_x.dot(theta.T)
```

```python
from sklearn.metrics import r2_score
print('R2: ',r2_score(pred_y,y))
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
#stochastic Gradient Descent
def SGD(x,y,learning_rate,n_epochs):
    Loss=[]
    theta0=[]
    theta1=[]
    lines=[]
    theta=np.random.rand(1,x.shape[1])
    for i in range(n_epochs):
        old_theta=theta.copy()
        print('--------------------epoch(',(i+1),')-----------------------')
        print(theta)
        for it in range(len(x)):
            index=it
            theta-=learning_rate*(((x[index].dot(theta.T)-y[index].reshape(-1,1))*x[index]))
            Loss.append(((x.dot(theta.T)-y.reshape(-1,1))**2).mean(axis=0))
            theta0.append(theta[0][0])
            theta1.append(theta[0][1])
            lines.append(x.dot(theta.T))
        check=theta==old_theta
        if check.all():
            break
    return theta,Loss,theta0,theta1,lines
theta,loss,theta0,theta1,lines=SGD(new_x,y,0.001,500)
```
